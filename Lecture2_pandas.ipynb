```python
import pandas as pd
import numpy as np

data = {
    'A': [10, 20, 30, 40],
    'B': [5, 15, 25, 35]
}

df = pd.DataFrame(data)

# .agg() მეთოდი საშუალებას გაძლევთ ერთდროულად შეასრულოთ რამდენიმე სტატისტიკური გამოთვლა თქვენს DataFrame-ზე.

# მრავლობითი აგრეგირება მონაცემთა ჩარჩოზე
aggregations = df.agg(['mean', 'std', 'min', 'max'])

# განსხვავებული აგრეგაციები სხვადასხვა სვეტზე
column_aggregations = df.agg({
    'A': ['min', 'max'],
    'B': ['mean', 'std']
})

# print("\nAggregations:\n", aggregations)
# print("\nColumn-wise Aggregations:\n", column_aggregations)


# cxrili 'Group' სვეტით
df_grouped = pd.DataFrame({
'Group': ['X', 'X', 'Y', 'Y'],
'Data': [1, 2, 3, 4]
})
print(df_grouped)
# დაჯგუფება 'Group' სვეტის მიხედვით და საშუალოს გამოთვლა
group_means = df_grouped.groupby('Group').mean()
# print("\nჯგუფების საშუალო მნიშვნელობები:\n", group_means)


# Calculate mean of each column
mean_values = df.mean()

# Sum of each row
row_sum = df.sum(axis=1)

# Descriptive statistics for each column
descriptive_stats = df.describe()

print("Mean values:\n", mean_values)
print("\nSum of rows:\n", row_sum)
print("\nDescriptive statistics:\n", descriptive_stats)

```
